## Indictments

```python
#read in data
df2 = pd.read_csv('DPD/cops_grand_jury.csv')
```

```python
#Convert relevant columns to datetimes
df2['Offense Date'] = pd.to_datetime(df2['Offense Date'])
df2['GJ Disposition Date'] = pd.to_datetime(df2['GJ Disposition Date'])
```

```python
#Create ‘Filed Year’ feature
df2['Filed Year'] = df2['Case Number'].str[1:3]
df2['Filed Year'] = '20' + df2['Filed Year']
df2['Filed Year'] = pd.to_numeric(df2['Filed Year'], errors='coerce')
```

```python
#All unique cases filed in 2019 and indicted in 2022
len(df2.loc[((df2['Filed Year'] <= 2019) & 
    (df2['GJ Disposition Date'] >= datetime(2022, 1, 1)) & 
    (df2['GJ Disposition'] == 'True Bill'))].groupby('Case Number').first())
```

```python
#All unique cases filed in 2019 and indicted in 2022
len(df2.loc[((df2['Filed Year'] <= 2018) & 
            (df2['GJ Disposition Date'] >= datetime(2021, 1, 1)) & 
            (df2['GJ Disposition Date'] < datetime(2022, 1, 1)) & 
            (df2['GJ Disposition'] == 'True Bill'))].groupby('Case Number').first())
```
